```python
# Import the numpy module as np and the matplotlib.pyplot package as plt.
# Use np.linspace to create a grid of 50 points ranging from 0 to 1.
# Create a numpy array
x = np.linspace(0,1,50)
print(x,'\n')
```

```python
#def log_fixed(arr):
#  with np.errstate(divide='ignore', invalid='ignore'):
#    result = np.log(arr)
#  return result

np.seterr(divide = 'ignore')

# containing the values for the natural logarithm function using the np.log(x) function. Create a numpy array
logarithm = np.log(x)

# containing the values for the exponential function using the np.exp(x) function.
exponential = np.exp(x)

# Use the plt.scatter(x,y) method for the vectors to create two scatter plots of the points you've created.
plt.scatter(x, logarithm, label = 'Natural Log', color = "orange")
plt.scatter(x, exponential, label ='Exponential', color='red')
# Use the plt.show() method to create the plot.


#add labels to the axes using plt.xlabel(label) and plt.ylabel(label). Add a title to the graph using plt.title(title), like "Natural Log and Exponential Functions".
plt.xlabel('x')
plt.ylabel('y')
plt.title('Natural Log and Exponential Functions')

#That looks a lot better, but we need a legend. When you screate the scatter plots, add the argument label='Natural Log' or label='Exponential' to the .scatter method call.
#Before the plt.show() call, add a plt.legend(loc = 'lower right') method call, which creates a legend in the lower right.

plt.legend(loc = 'lower right')

plt.grid(True)
plt.show()
```
