Use the reshaped X_test and y_test data to create a k-nearest neighbor classifier of digit. What is the optimal number of neighbors k? If you can't determine this, play around with different values of k for your classifier.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Reshape X_train and X_test into 2D arrays (n_samples, n_features)
X_train = X_train.reshape((X_train.shape[0], -1))  # (60000, 784)
X_test = X_test.reshape((X_test.shape[0], -1))     # (10000, 784)

# Initialize KNN classifier (k=3 is a common default)
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier on the training data
knn.fit(X_train, y_train)

# Make predictions on the test data
y_pred = knn.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of KNN on test data: {accuracy:.2f}")
```

```python

# Range of k values to try
k_values = range(1, 51)
accuracies = []

# Loop over different values of k
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    # Calculate and store accuracy for current k
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Plot the relationship between k and accuracy
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o')
plt.title('K vs. Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()

# Find the k with the highest accuracy
optimal_k = k_values[accuracies.index(max(accuracies))]
print(f"The optimal number of neighbors is k={optimal_k} with accuracy={max(accuracies):.2f}")
```

For the optimal number of neighbors, how well does your predictor perform on the test set? Use a confusion matrix and compute accuracy.

```python
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

# Train KNN with the optimal number of neighbors (k=3)
knn_optimal = KNeighborsClassifier(n_neighbors=3)
knn_optimal.fit(X_train, y_train)

# Predict on the test set
y_pred_optimal = knn_optimal.predict(X_test)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred_optimal)

# Load the digits dataset
digits = load_digits()

# Use the same confusion matrix plotting code
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.title("Confusion Matrix for k=3")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Compute accuracy
accuracy = accuracy_score(y_test, y_pred_optimal)
print(f"Accuracy of KNN on test data with k=3: {accuracy:.2f}")
```
